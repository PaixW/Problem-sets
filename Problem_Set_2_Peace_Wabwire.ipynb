```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Tweets.csv",on_bad_lines='skip')
```

```python
unique_sentiment = np.sort(df["airline_sentiment"].unique())
print(unique_sentiment)

_ = plt.figure()
plt.hist(df["airline_sentiment"])
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```
